# EDA using Bivariate Analysis

```python
tips=sns.load_dataset('tips')
```

```python
titanic=pd.read_csv('train.csv')
```

```python
flights=sns.load_dataset('flights')
```

```python
iris=pd.read_csv('iris.csv')
```

```python
tips.head()
```

```python
flights.head()
```

```python
#scatterplot(Numerical-Numerical)
sns.scatterplot(x=tips['total_bill'],y=tips[ 'tip'], hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
#Bar plot(num-cat)
titanic.head()
```

```python
sns.barplot(data=titanic,x=titanic['Pclass'],y='Age',hue='Sex')
```

```python
#BOX plot(num-cat)
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

```python
#Distplot(num-cat)
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
#Heatmap(cat-cat)
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100)
```

```python
#Clustermap(cat-cat)
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
#pairplot
iris.head()
```

```python
sns.pairplot(iris,hue='Species')
```

```python
#lineplot(num-num)
flights.head()
```

```python
new=flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
